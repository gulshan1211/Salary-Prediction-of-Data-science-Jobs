```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
```

```python
df.job_simp.value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

```python
df['job_state']= df.job_state.apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

```python
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

```python
df['num_comp'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != '-1' else 0)
```

```python
df['Competitors']
```

```python
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)
```

```python
df[df.hourly ==1][['hourly','min_salary','max_salary']]
```

```python
df['company_txt'] = df.company_txt.apply(lambda x: x.replace('\n', ''))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df[['age', 'avg_salary', 'Rating', 'desc_len', 'num_comp']].corr()

cmap = sns.diverging_palette(220, 10, as_cmap=True)

plt.figure(figsize=(5, 5))  
sns.heatmap(corr_matrix, vmin=-1, vmax=1, center=0, cmap=cmap, annot=True,
            fmt=".2f", square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title("Correlation Heatmap")
plt.xticks(rotation=45, ha='right') 
plt.yticks(rotation=0)  

plt.tight_layout()  

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cov_matrix = df[['age', 'avg_salary', 'Rating', 'desc_len', 'num_comp']].cov()

cmap = sns.diverging_palette(220, 10, as_cmap=True)

plt.figure(figsize=(6, 6))
ax = sns.heatmap(cov_matrix, cmap=cmap, annot=True, fmt=".2f", square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title("Covariance Heatmap")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

ax.tick_params(axis='both', labelsize=10) 

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_cat = df[['Location', 'Headquarters', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'company_txt', 'job_state', 'same_state', 'python_yn', 'R_yn', 'spark', 'aws', 'excel', 'job_simp', 'seniority']]

custom_palette = sns.color_palette('pastel')

for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    num_categories = len(cat_num)

    if num_categories <= 25:
        plt.figure(figsize=(4, 6), facecolor='white') 

        chart = sns.barplot(x=cat_num.index, y=cat_num, palette=custom_palette, linewidth=0.2)
        chart.set_xticklabels(chart.get_xticklabels(), rotation=90)

        plt.xlabel(i)
        plt.ylabel('Count')
        plt.title('Bar Chart for {}'.format(i))
        plt.tight_layout()

        plt.show()
    else:
        print("Graph for %s not generated: Total categories = %d" % (i, num_categories))
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
